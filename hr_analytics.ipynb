```python
#required libraries
import pandas as pd
import matplotlib.pyplot as plt

# loading the dataset
df = pd.read_csv("HR_Analytics.csv.xls")

```

```python
df.head()
```

```python
#checking basic information
print("\nDataset Information:\n")
print(df.info())
# info() prints : total rows nd columns,datatypes of columns,non-null values
```

```python
#checking missing values
print("\nMissing Values\n")
print(df.isnull().sum())
```

```python
#Checking duplicate rows
print("\nDuplicate Rows:\n")
print(df.duplicated().sum())
```

```python
#Statistical Summary
print("\nStatistical Sumaary\n")
print(df.describe())

#describes() prints: mean,minimum value,maximum value nd standard deviation 
```

```python
#Attrition Analysis
print("Attrition Count:\n")
attrition=df['Attrition'].value_counts()
print(attrition)
```

```python
#Department Analysis
print("Average Monthly Income based on Job Satisfaction:\n")
satisfaction=df.groupby('JobSatisfaction')['MonthlyIncome'].mean()
print(satisfaction)

#Checking Whether Higher Satisfaction is Related to better income
```

```python
#Age distribution chart
df['Age'].plot(kind='hist')
plt.title("Employe Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Employes")
plt.show()        #histogram chart
```

```python
#Attrition chart
attrition.plot(kind='bar')
plt.title("Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Employees")
plt.show()     #bar chart
```

```python
dept=df['Department'].value_counts()

print(dept)
```

```python
#Department chart
dept.plot(kind='bar')
plt.title("Employee by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

```python
#Monthly Income analysis
income=df.groupby('Department')['MonthlyIncome'].mean()
income.plot(kind='bar')
plt.title("Average Income Department")
plt.xlabel("Department")
plt.ylabel("Average Income")
plt.show()
```

```python
#Create new column AgeGroup
labels=['Young Adult','Adult','Middle-aged','Senior']
df['AgeGroup']=pd.qcut(df['Age'],q=4,labels=labels)
```
